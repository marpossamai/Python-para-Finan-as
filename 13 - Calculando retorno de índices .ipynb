```python
import numpy as np
import pandas as pd
from pandas_datareader import data as web
import matplotlib.pyplot as plt
```

```python
tickers = ['^GSPC', '^IXIC', '^GDAXI', '^DJI']
ind_data = pd.DataFrame()

for t in tickers:
    ind_data[t] = web.DataReader(t, data_source='yahoo', start='1997-1-1', end='2017-1-1')['Adj Close']
```

```python
ind_data.info()
```

```python
ind_data.head()
```

```python
ind_data.tail()
```

```python
(ind_data / ind_data.iloc[0] * 100).plot(figsize=(15, 6));
plt.show()
```

```python
retorno_ind = (ind_data / ind_data.shift(1)) - 1
retorno_ind.tail()
```

```python
retorno_ind_anual = retorno_ind.mean() * 250
retorno_ind_anual
```
